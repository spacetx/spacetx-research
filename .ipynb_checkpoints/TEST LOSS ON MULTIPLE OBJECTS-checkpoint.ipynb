```python
class DecoderConv(nn.Module):
    """ Decode z -> x
        INPUT:  z of shape: batch x dim_z
        OUTPUT: image of shape: batch x ch_out x width x height (where width = height = 28)
    """

    def __init__(self, params: dict, dim_z=None, ch_out=None):
        super().__init__()
        self.dim_z = dim_z
        self.ch_out = ch_out
        self.width = params['SD.width']
        assert self.width == 28  # the following architecture assume width=height=28

        self.upsample = nn.Linear(self.dim_z, 64 * 7 * 7)
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(64, 32, 4, 2, 1),  # B,  64,  14,  14
            nn.ReLU(),
            nn.ConvTranspose2d(32, 32, 4, 2, 1, 1),  # B,  32, 28, 28
            nn.ReLU(),
            nn.ConvTranspose2d(32, self.ch_out, 4, 1, 2)  # B, 2*ch_out, 28, 28
        )

    def forward(self, z):
        independent_dim = list(z.shape[:-1])  # this includes: enumeration dimension, n_boxes, batch

        x1 = self.upsample(z).view(-1, 64, 7, 7)
        x2 = self.decoder(x1).view(independent_dim + [self.ch_out, self.width, self.width])
        return torch.sigmoid(x2)
    
def make_red(w):
    assert w.shape[-3] == 1  # check there is only one channel
    if len(w.shape) == 4:
        x = w.repeat(1,3,1,1)
    if len(w.shape) == 5:
        x = w.repeat(1,1,3,1,1)
    x[...,1:,:,:]=0
    return x

def draw_bb_only(z_where=None, width=None, height=None):
    with torch.no_grad():

        # Extract the probabilities for each box
        boxes, batch_size = z_where.bx.shape[:2]

        # prepare the storage
        batch_bb_np = np.zeros((batch_size, width, height, 3))  # numpy storage for bounding box images

        # compute the coordinates of the bounding boxes and the probability of each box
        x1 = (z_where.bx - 0.5 * z_where.bw).view(boxes, batch_size, 1)
        x3 = (z_where.bx + 0.5 * z_where.bw).view(boxes, batch_size, 1)
        y1 = (z_where.by - 0.5 * z_where.bh).view(boxes, batch_size, 1)
        y3 = (z_where.by + 0.5 * z_where.bh).view(boxes, batch_size, 1)
        x1y1x3y3 = torch.cat((x1, y1, x3, y3), dim=-1) 

        # draw the bounding boxes
        for batch in range(batch_size):

            # Draw on PIL
            img = Image.new('RGB', (width, height), color=0)
            draw = ImageDraw.Draw(img)
            for n in range(boxes):
                draw.rectangle(x1y1x3y3[n,batch, :].cpu().numpy(), outline='red', fill='red')
            batch_bb_np[batch, ...] = np.array(img.getdata(), np.uint8).reshape(width, height, int(3))

        # Transform np to torch, rescale from [0,255] to (0,1)
        batch_bb_torch = torch.from_numpy(batch_bb_np).permute(0, 3, 2, 1).float() / 255  # permute(0,3,2,1) is CORRECT
        return batch_bb_torch.to(z_where.bx.device)
    
def make_image(params=None, n_imgs=10, n_boxes=3):
    w = params['IMG.size_raw_image'] 
    ch = len(params['IMG.ch_in_description'])
    min_size = params['IMG.size_object_min']
    max_size = params['IMG.size_object_max']
    small_width = params['SD.width']
    threshold_overlap = params['NMS.overlap_threshold']
    
    uncropper = Uncropper()
    
    # make a small disk
    x_mat = torch.linspace(-1.0, 1.0,steps=small_width).view(-1,1).expand(small_width,small_width).float()
    y_mat = x_mat.clone().permute(1,0)
    r2   = (x_mat**2 + y_mat**2)
    disk = (r2<1.0).float()
    small_disks = disk[None,None,None,...].expand(n_boxes,n_imgs,-1,-1,-1) 
    small_squares = torch.ones_like(small_disks)
    small_stuff = torch.cat((small_disks,small_squares),dim=-3) # cut along the channel dimension
    
    # coordinate of object
    ix = torch.randint(low=1, high=w-max_size, size=[n_boxes,n_imgs])
    iy = torch.randint(low=0, high=w-max_size, size=[n_boxes,n_imgs])
    dx = torch.randint(low=min_size, high=max_size, size=[n_boxes,n_imgs])
    dy = torch.randint(low=min_size, high=max_size, size=[n_boxes,n_imgs])
    
    # Exact stuff
    bx = ix.float()+ 0.5*dx.float() 
    by = iy.float()+ 0.5*dy.float() 
    bw = dx.float() 
    bh = dy.float() 
    z_where_exact = collections.namedtuple("zwhere","bx by bw bh")._make([bx,by,bw,bh])
    
    # Approximate stuff
    bx = torch.clamp(bx + 0.5*min_size*torch.randn(bx.shape), min=0, max=w)
    by = torch.clamp(by + 0.5*min_size*torch.randn(by.shape), min=0, max=w)
    bw = torch.clamp(bw + max_size*torch.rand(bw.shape), min=min_size, max=max_size)
    bh = torch.clamp(bh + max_size*torch.rand(bh.shape), min=min_size, max=max_size)
    z_where_noise = collections.namedtuple("zwhere","bx by bw bh")._make([bx,by,bw,bh])
    
    # Make the big disks
    big_stuff = uncropper.forward(z_where_exact, small_stuff, w, w)
    big_disks, big_squares = torch.split(big_stuff,[1,1],dim=-3)
    
    # compute the intersection over min_area of the disks
    flatten = big_disks.view(n_boxes,n_imgs,-1)
    area = torch.sum(flatten,dim=-1)
    f1 = flatten.unsqueeze(0)
    f2 = flatten.unsqueeze(1)
    intersection_matrix = torch.sum(f1*f2,dim=-1)
    min_area_matrix = torch.min(area.unsqueeze(0),area.unsqueeze(1))
    IoMIN = intersection_matrix/min_area_matrix
    cluster_mask = (IoMIN > threshold_overlap).float()
    score = torch.sigmoid(torch.rand(n_boxes,n_imgs))
    possible= torch.ones_like(score)
    selected = NmsBaseClass.perform_nms_selection(cluster_mask,score,possible,n_boxes)
    
    # add only the disk which survive NMS
    #print("selected.shape",selected.shape)
    #print("big_disks.shape",big_disks.shape)
    sum_disks = torch.sum(big_disks*selected[...,None,None,None],dim=-5)
    #print("sum_disks.shape",sum_disks.shape)

    # add the background
    background = 0.2+0.1*torch.randn(sum_disks.shape)
    imgs = sum_disks + (1 - sum_disks)*background
    
    return imgs, z_where_exact, z_where_noise
```

```python
hyper_params = SimulationDictionary()
hyper_params["size_object_min"]=10
hyper_params["NMS.overlap_threshold"]=0.5
hyper_params["IMG.n_objects_max"]=3
#print(hyper_params)
imgs_input, z_where_exact, z_where_noise = make_image(params=hyper_params, 
                                                      n_imgs=20, 
                                                      n_boxes=hyper_params["IMG.n_objects_max"])
bb_exact = draw_bb_only(z_where = z_where_exact,
                  width = imgs_input.shape[-2],
                  height = imgs_input.shape[-1])
bb_noise = draw_bb_only(z_where = z_where_noise,
                  width = imgs_input.shape[-2],
                  height = imgs_input.shape[-1])
```

```python
class ComplexVae(nn.Module):
        
    def __init__(self, params):
        super().__init__()

        # Instantiate all the modules
        self.uncropper = Uncropper()
        self.mask_decoder = DecoderConv(params, dim_z = 10, ch_out = 1)

        self.size_min = params['IMG.size_object_min']
        self.size_max = params['IMG.size_object_max']
        self.size_raw_image = params['IMG.size_raw_image']
        self.ch_in_description = params['IMG.ch_in_description']
        self.ch_raw_image = len(self.ch_in_description)
        self.small_width = params['SD.width']
        self.n_objects_max  = params["IMG.n_objects_max"]
        
        # losses 
        self.logp_normal = params['LOSS.logp_normal']
        self.logp_cauchy = params['LOSS.logp_cauchy']

        # Put everything on the cuda if necessary
        self.use_cuda = params["use_cuda"]
        if self.use_cuda:
            self.cuda()
            
    def guide(self, imgs=None, z_where_input=None, z_where_inferred=None, z_mask_inferred=None, verbose=False):
        # Trick
        if imgs is None:
            imgs = torch.zeros(8, self.ch_raw_image, self.size_raw_image, self.size_raw_image)
            if self.use_cuda:
                imgs = imgs.cuda()
        assert len(imgs.shape) == 4
        batch_size, ch, width, height = imgs.shape
        assert width == height
        one = torch.ones([1])
        # Enf of Trick

        batch_plate = pyro.plate("batch", batch_size, dim=-1)
        box_plate = pyro.plate("boxes", self.n_objects_max, dim=-2)
        
        
        # Define the parameters
        z_mask_mu = pyro.param("z_mask_mu", 
                                dist.Normal(0,one).expand([self.n_objects_max, batch_size, 10]).sample(), 
                                constraint=constraints.interval(0.1, 1.0))
        z_mask_std = pyro.param("z_mask_std", torch.ones_like(z_mask_mu), 
                                constraint=constraints.positive)
        bx_mu = pyro.param("bx_mu", z_where_input.bx, 
                            constraint=constraints.interval(0, width))
        by_mu = pyro.param("by_mu", z_where_input.by, 
                            constraint=constraints.interval(0, width))
        bw_mu = pyro.param("bw_mu", z_where_input.bw, 
                            constraint=constraints.interval(self.size_min,self.size_max))
        bh_mu = pyro.param("bh_mu", z_where_input.bh, 
                            constraint=constraints.interval(self.size_min,self.size_max))                        
        bx_std = pyro.param("bx_std", 10*torch.ones_like(bx_mu), 
                            constraint=constraints.interval(0.1, self.size_min))
        by_std = pyro.param("by_std", 10*torch.ones_like(bx_mu), 
                            constraint=constraints.interval(0.1, self.size_min))
        bw_std = pyro.param("bw_std", 10*torch.ones_like(bx_mu), 
                            constraint=constraints.interval(0.1, self.size_min))
        bh_std = pyro.param("bh_std", 10*torch.ones_like(bx_mu), 
                            constraint=constraints.interval(0.1, self.size_min))
        logit_mu = pyro.param("logit_mu", torch.zeros_like(bx_mu), constraint=constraints.real)
        logit_std = pyro.param("logit_std", torch.ones_like(bx_mu), constraint=constraints.interval(0.1, 1.0))            
        
        with batch_plate, box_plate:
            
            # sample 
            logit = pyro.sample("logit", dist.Normal(logit_mu,logit_std))
                            
            # z_where
            bx = pyro.sample("bx", dist.Normal(bx_mu,bx_std))
            by = pyro.sample("by", dist.Normal(by_mu,by_std))
            bw = pyro.sample("bw", dist.Normal(bw_mu,bw_std))
            bh = pyro.sample("bh", dist.Normal(bh_mu,bh_std))
                         
            # z_mask
            z_mask = pyro.sample("z_mask", dist.Normal(z_mask_mu,z_mask_std).to_event(1))
                
            #print("in guide")
            #print("z_mask.shape",z_mask.shape)
            #print("bx.shape",bx.shape)
            #print("logit.shape",logit.shape)                
            
    def model(self, 
              imgs=None, 
              z_where_input=None, 
              z_where_inferred=None, 
              z_mask_inferred=None, 
              logit_inferred=None, 
              verbose=False):
        # Trick
        if imgs is None:
            observed = False
            imgs = torch.zeros(20, self.ch_raw_image, self.size_raw_image, self.size_raw_image)
            if self.use_cuda:
                imgs = imgs.cuda()
        else:
            observed = True
        assert len(imgs.shape) == 4
        batch_size, ch, width, height = imgs.shape
        assert width == height
        # Enf of Trick

        # register the modules 
        pyro.module("uncropper", self.uncropper)
        pyro.module("mask_decoder", self.mask_decoder)

        # Parameters
        zero = torch.zeros(1, dtype=imgs.dtype, device=imgs.device)
        one = torch.ones(1, dtype=imgs.dtype, device=imgs.device)
        one_ch = torch.ones(ch, dtype=imgs.dtype, device=imgs.device)
        if getattr(self, "detach_parameters", False):
            bg_mu = 0.2*one_ch.view(-1,1,1)
            fg_mu = 0.8*one_ch.view(-1,1,1)
            bg_std = 0.1*one_ch.view(-1,1,1)
            fg_std = 0.1*one_ch.view(-1,1,1)
            normal_sigma = 0.1*one_ch.view(-1,1,1)
        else:
            bg_mu = pyro.param("background_mu", 0.2*one_ch, constraint=constraints.unit_interval).view(-1, 1, 1)
            fg_mu = pyro.param("foreground_mu", 0.8*one_ch, constraint=constraints.unit_interval).view(-1, 1, 1)
            bg_std = pyro.param("background_std", 0.2*one_ch, constraint=constraints.interval(0.01, 0.50)).view(-1, 1, 1)
            fg_std = pyro.param("foreground_std", 0.2*one_ch, constraint=constraints.interval(0.01, 0.50)).view(-1, 1, 1)
            normal_sigma = pyro.param("normal_std", 0.2*one_ch, constraint=constraints.interval(0.01, 0.50)).view(-1, 1, 1)

        batch_plate = pyro.plate("batch", batch_size, dim=-1)
        box_plate = pyro.plate("boxes", self.n_objects_max, dim=-2)
        
        with batch_plate, box_plate:
        
            if z_where_inferred is None:
                bx = pyro.sample("bx", UniformWithTails(zero,width,self.size_min))
                by = pyro.sample("by", UniformWithTails(zero,width,self.size_min))
                bw = pyro.sample("bw", UniformWithTails(one*self.size_min,self.size_max,0.2*self.size_min))
                bh = pyro.sample("bh", UniformWithTails(one*self.size_min,self.size_max,0.2*self.size_min))
                z_where=collections.namedtuple("z_where","bx by bw bh")._make([bx,by,bw,bh])      
            else:
                z_where = z_where_inferred
                
            if logit_inferred is None:
                logit = pyro.sample("logit", dist.Normal(zero,1))
            else:
                logit = logit_inferred
                    
            if z_mask_inferred is None:
                z_mask = pyro.sample("z_mask", dist.Normal(zero,1).expand([10]).to_event(1))
            else:
                z_mask = z_mask_inferred
                    
            p = torch.sigmoid(logit)
                         
            #print("in model")
            #print("z_mask.shape",z_mask.shape)
            #print("bx.shape",z_where.bx.shape)
            #print("logit.shape",logit.shape)          
                    
        with  batch_plate:
            
            # Generation of images
            small_mu = torch.ones(self.n_objects_max,batch_size,ch,self.small_width,self.small_width)
            small_w = self.mask_decoder(z_mask)
            assert small_w.shape == small_mu.shape
            small_squares = torch.ones_like(small_mu)
            small_stuff = torch.cat((small_mu, small_w, small_squares), dim=-3)  # cat ch dimension
            ch_dims = (small_mu.shape[-3], small_w.shape[-3], small_squares.shape[-3])
            big_stuff = self.uncropper.forward(z_where, small_stuff, width, height)
            big_mu, big_w, big_squares = torch.split(big_stuff, ch_dims, dim=-3)  # split ch dimension
            def_bg_mask = 1.0 - big_mu
            #print("small_mu.shape",small_mu.shape)
            #print("small_w.shape",small_w.shape)
            #print("small_squares.shape",small_squares.shape)
            #print("big_mu.shape",big_mu.shape)
            #print("big_w.shape",big_w.shape)
            #print("big_squares.shape",big_squares.shape)
            
            if not observed:
                
                logp_tot = None
            
            else: 
                
                # Cost if out of range
                # This is necessary b/c box out of range does not take any other gradient
                x1 = z_where.bx - 0.5*z_where.bw
                x3 = z_where.bx + 0.5*z_where.bw
                y1 = z_where.by - 0.5*z_where.bh
                y3 = z_where.by + 0.5*z_where.bh
                cost_out_of_range = torch.clamp(-x1,min=0) + \
                                    torch.clamp(x3-width, min=0) + \
                                    torch.clamp(-y1,min=0) + \
                                    torch.clamp(y3-height, min=0)
                
                # Cost if volume of the mask is too much or too tittle.
                # Otherwise the mask becomes zero and the model is ill defined
                volume_box = (z_where.bw * z_where.bh).squeeze(-1)
                volume_mask = torch.sum(big_w, dim=(-1, -2, -3))
                cost_volume_fraction = torch.clamp(self.min_of * volume_box - volume_mask, min=0)
                cost_volume_absolute = torch.clamp(volume_mask - self.volume_mask_max, min=0) +\
                                       torch.clamp(self.volume_mask_min - volume_mask, min=0)
                
                # Cost if mask does not touch the edge of the box
                # Otherwise the bounding box is completely detached from the mask
                left = (torch.sum(small_w[...,0,:2,:], dim=(-1,-2))<0.1).float()
                right = (torch.sum(small_w[...,0,-2:,:], dim=(-1,-2))<0.1).float()
                bottom = (torch.sum(small_w[...,0,:,:2], dim=(-1,-2))<0.1).float()
                top = (torch.sum(small_w[...,0,:,-2:], dim=(-1,-2))<0.1).float()
                # these terms do (respectively):
                # reduce bw, reduce bh, reduce bx, increase bx, reduce by, increase by
                cost_not_fitted = z_where.bw*left*right + \
                                  z_where.bh*top*bottom + \
                                  z_where.bx*(1-left)*right + \
                                  z_where.bx*left*(right-1) + \
                                  z_where.by*(1-bottom)*top + \
                                  z_where.by*bottom*(top-1)
                
                # Cost if masks overlap:
                # cost_i = 0.5 * p_i * M_i * [ \sum_{j \ne i} p_j M_j ] 
                #        = 0.5 * p_i * M_i * (K - p_i * M_i) where 
                # K = \sum_j p_j M_j
                p_times_MASK = big_w * p[...,None,None,None]  # add singleton for ch,w,h
                K = torch.sum(p_times_MASK, dim=-5, keepdims=True)  # sum over boxes
                cost_overlap = 0.5 * torch.sum(p_times_MASK * (K - p_times_MASK), dim=(-1,-2,-3))  # sum over ch,w,h 
                
                # put all the cost together
                costs = self.lambda_volume_fraction*cost_volume_fraction + \
                        self.lambda_not_fitted*cost_not_fitted + \
                        self.lambda_out_of_range*cost_out_of_range + \
                        self.lambda_overlap*cost_overlap 
                #print("costs.shape",costs.shape)
                
                if(verbose):
                    print("too_small",-self.lambda_volume_fraction*cost_volume_fraction) 
                    print("out_of_range",-self.lambda_out_of_range*cost_out_of_range)
                    print("not_fitted",-self.lambda_not_fitted*cost_not_fitted)
                    print("overlap",-self.lambda_overlap*cost_overlap)
                    
                    #print("debug 1",torch.clamp(-x1,min=0))
                    #print("debug 2",torch.clamp(x3-width, min=0))
                    #print("debug 3",torch.clamp(-y1,min=0))
                    #print("debug 4",torch.clamp(y3-height, min=0))
                    
                    #print("debug a",z_where.bw*left*right)
                    #print("debug b",z_where.bh*top*bottom)
                    #print("debug c",z_where.bx*(1-left)*right)
                    #print("debug d",z_where.bx*left*(right-1))
                    #print("debug e",z_where.by*(1-bottom)*top)
                    #print("debug f",z_where.by*bottom*(top-1))
                
                # objective function
                logp_off = UnitCauchy(bg_mu, bg_std).log_prob(imgs)
                logp_on_cauchy = UnitCauchy(fg_mu, fg_std).log_prob(imgs)
                logp_on_normal = UnitCauchy(big_mu, normal_sigma).log_prob(imgs)

                rec_all_boxes = torch.sum(p_times_MASK * (logp_on_normal - logp_off), dim=(-1,-2,-3))
                rec = torch.sum(logp_off, dim=(-1,-2,-3)) + rec_all_boxes 
                
                # Note that I am summing over the boxes
                logp_tot = torch.sum(rec - p.detach() * costs, dim=-2)  # note that the costs do no change p
                assert len(logp_tot.shape) == 1
                assert logp_tot.shape[0] == batch_size 

                pyro.sample("LOSS", CustomLogProbTerm(custom_log_probs=logp_tot), obs=logp_tot)
            
            return logp_tot, small_mu, small_w, small_squares, big_mu, big_w, big_squares, z_where, z_mask, logit


    def save_zwhere(self):
        bx_mu = pyro.param("bx_mu")
        by_mu = pyro.param("by_mu")
        bw_mu = pyro.param("bw_mu")
        bh_mu = pyro.param("bh_mu")
        z_where=collections.namedtuple("z_where","bx by bw bh")._make([bx_mu,by_mu,bw_mu,bh_mu])
        return z_where
    
    def save_zmask(self):
        z_mask = pyro.param("z_mask_mu")
        return z_mask

    def save_logit(self):
        logit = pyro.param("logit_mu")
        return logit
    
    def reconstruct_img(self, imgs, z_where=None, z_mask=None):
        
        if z_where is None:
            z_where = self.save_zwhere()
        if z_mask is None:
            z_mask = self.save_zmask()

        if self.use_cuda:
            imgs = imgs.cuda()

        with torch.no_grad():  # do not keep track of the gradients
            self.eval()  # set the model into evaluation mode
            return vae.model(z_where_inferred=z_where, z_mask_inferred=z_mask)
```

```python
param_store = pyro.get_param_store()
param_store.clear()

vae = ComplexVae(hyper_params)
optimizer =  Adam({"lr": 1.0e-2})
train_loss = []
test_loss = []
history_dict = {
    "logit" : [],
    "zwhere" : [],
    "zmask" : [],
    
    "train_loss" : [],
    "test_loss" : []
    }

history_dict["zwhere"].append(z_where_exact)
history_dict["zwhere"].append(z_where_noise)
history_dict["zmask"].append(None)
history_dict["zmask"].append(None)
history_dict["logit"].append(None)
history_dict["logit"].append(None)

svi = SVI(vae.model, vae.guide, optimizer, loss=Trace_ELBO(max_plate_nesting=2))
```

```python
NUM_EPOCHS=2000

for delta_epoch in range(0,NUM_EPOCHS+1):
    
    vae.detach_parameters = True
    vae.logp_on_normal = 1.0
    vae.logp_on_cauchy = 0.0
    
    # regularization absolute volume mask
    vae.volume_mask_max = 1
    vae.volume_mask_min = 1
    vae.lambda_volume_absolute = 0
    
    # regularization box out of range 
    # large because single term acting on box
    # It should be extensive
    vae.lambda_out_of_range = 100
    
    # regularization not fitted 
    # large because single term acting on box
    # It should be extensive
    vae.lambda_not_fitted = 100.0
    
    # regularization fractional volume mask
    # small because it acts on all pixels
    # It should be intensive
    vae.min_of = 0.1
    vae.lambda_volume_fraction = 1.0
    
    # regularization overlap
    vae.lambda_overlap = 0.0
    
    with torch.autograd.set_detect_anomaly(True):
        loss = svi.step(imgs=imgs_input, z_where_input=z_where_noise, verbose=((delta_epoch%100)== 0))
        
    history_dict["train_loss"].append(loss)
    if(delta_epoch % 100) ==0:
        print("epoch loss",delta_epoch,loss)  
        z_where = vae.save_zwhere()
        z_mask = vae.save_zmask()
        logit = vae.save_logit()
        history_dict["zwhere"].append(z_where)
        history_dict["zmask"].append(z_mask)
        history_dict["logit"].append(logit)
        history_dict["test_loss"].append(loss)
```
